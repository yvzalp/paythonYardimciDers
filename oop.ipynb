```python
#puplic,proctected ve private tanimlayicilar

class Sporcu():
    def __init__(self, isim ,branş,altin,gümüs,bronz):
        self.isim=isim
        self.branş=branş
        self.bronz=bronz #public
        self._gümüs=gümüs #yari gizli prockted tek alt tire ile
        self.__altin=altin #private tmm gizli clastan erişim var ancak bir methotla erişime acilablilr
        
    def yazdir(self):
        return "ismim {} ,branşım {} dir ".format(self.isim,self.branş)
    
    def a_erişim(self): # private olan altın eğişkenini erişime actik
        madalya=self.__altin
        return "sporcunun altın madalya sayisi {} ".format(madalya)
    
    
atlet=Sporcu("ahmed","atletizim",1,3,9) 
#print(atlet.branş,atlet.isim)
#atlet.yazdir()

print(atlet.bronz)
print(atlet._gümüs)
#print(atlet.__altin)

atlet.a_erişim()


  
```

```python
#class method,static methot
class Calisan:
    
    zam_orani=1.05                                                          #code calişiramadik
   
    def __init__(self,ad,soyad,maas):
        self.ad=ad
        self.soyad=soyad
        self.maas=maas
        self.email=self.ad+self.soyad+"@gmail.com"
        pass
    def zamla(self):
        #self.maas=(self.maas*1.05) #burda clas variable kullanmadik.dinamik degil
        self.maas=(self.maas*self.zam_orani) #1.yöntem dinamik self ile
        #self.maas=(self.maas*Calisan.zam_orani)#2.yöntem dinamik Class ismi ile
        #self.maas=(self.maas*zam_orani) -->bu hali hata olur değişkenin scop alan degil
        
    @classmethod                # burda class seviyeside bir method yaptik hem örnek hem sınıf kapsiyor
    def degistir(cls,yeni):
        cls.zam_orani=yeni  
    
    @classmethod
    def yPersonel(cls,personelBilgi):
        ad,soyad,maas=personelBilgi.split(" ")
        return cls(ad,soyad,maas)
        

    @staticmethod  #static methot bagimsiz düşünebiliriz 
    def telNoDüzenle(mumara): #default parametresi yoktur
        return numara.split(" ")
        


Calisan.degistir(1.6) #bu sekilde bütün personele etki eder.class seviyesinde oldugu içn class isimi ile cagirdik

#personel 1
personel1=Calisan("ahmet","binici",5000)
personel1.zamla()                        #bu sekilde bir methotla sadece personel 1 maasişna etki ediyor 
print("personel 1 :", personel1.maas)

#mPersonel
mPersonel="mucteba binici 10000"    #metin girip parcalayip atama yaptik
# ad,soyad,maas=mPersonel.split(",")    #bu kısmı yukari metodun içine alip nesneden üretmeyi daha dinamik hale getirelim
# print(ad)
# mPersonel=Calisan(ad,soyad,maas) #dikkat edildiyse nesneden nesne üretmiş oluyoruz
# print(mPersonel.email)


#yPersonel
yPersonel=Calisan.yPersonel(mPersonel)  #burda yukard yaptiğimiz nesneden nesne üretmeyi dinamik hale getiriyoruz
print(yPersonel.email)


#tel ekleme static methot ile
tel="0539 458 25 30"
print(Calisan.telNoDüzenle(tel))






```

```python
#inharetens,Subclass,Parent,Child

class Calisan():# burd class'a method parantezi koyduk sebebi parent child ilişkisi kurabilmak için
    
    zam_orani=1.05
     
    
    def __init__(self,ad,soyad,maas):
        self.ad=ad
        self.soyad=soyad
        self.maas=maas
        self.email=self.ad+self.soyad+"@gmail.com"
        pass
    
    def tamAd(self):
        return "benim ismim {} sayadim {} maaşim ise {} ttl dir.".format(self.ad,self.soyad,self.maas)
    def zamla(self):
        
        self.maas=(self.maas*Calisan.zam_orani)
        
    





class Gelistirici(Calisan):
    def __init__(self,ad, soyad,maas,pytDili):
       self.pytDili=pytDili
       Calisan.zam_orani=1.5  #geliştiriciyi yen eklemştik şimdide zam yaptik -self- yazabiliriz
       #super().__init__(ad, soyad, maas)  #1.yol super de self yok
       Calisan.__init__(self,ad, soyad, maas)
       




class Yönetici(Calisan):
    def __init__(self, ad, soyad, maas,isciler=None):#burda yöneticiye calişan olusturacaz ama boş gelirse -None- olsun
        super().__init__(ad, soyad, maas)
        
        if isciler is None:
            self.isciler=[]
        else:
            self.isciler=isciler 
            
        def eleman_ekle(self,eleman):#eleman ekledik dişardan mudahle
            self.isciler.append(eleman)
            
        def eleman_cikar(self,eleman):#eleman cikardik dişardan mudahle
            self.isciler.remove(eleman) 
                    
        def isimListele(self):
            for eleman in self.isciler:
                print(eleman.tamAd())






 
gel1 =Gelistirici("Altay","ak",4000,"paython")  
print(gel1.ad)
gel1.zamla()
print(gel1.maas)


#personel 1
personel1=Calisan("ahmet","binici",5000)
#personel 2
personel2=Calisan("temel","arkın",10000)


                  


yönet=Yönetici("kazım","dogan",3000,[gel1])
print(yönet.isimListele())
yönet.eleman_ekle(personel1)    #eleman eklemek icin personel1 gönderdik ,ayni şekilde silme işlmeide yapabiliriz
print(yönet.isimListele())



#parent-child ilişkisi nasil ögreniriz

print(isinstance(Yönetici,Calisan)) #alt sınıf oldugunu dogrulanmak istersek ilk onu yazıyoruz
print(issubclass(Gelistirici,Calisan)) #2.yol olarak kullanilabilir



```
